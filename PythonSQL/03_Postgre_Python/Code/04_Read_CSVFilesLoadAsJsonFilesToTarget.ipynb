```python
import pandas as pd
import glob
import json
import re # regular expression
import os
```

```python
#get dir and sub directory list /**
#glob.glob('F:/GitHub/PythonDataEngineering/PythonSQL/03_Postgre_Python/data-master/retail_db/**',recursive=True)
##get only the files in all subfolders /*/*
# Also to filter the file name with specific pattern we /*/part**
src_file_names=glob.glob('F:/GitHub/PythonDataEngineering/PythonSQL/03_Postgre_Python/data-master/retail_db/*/part*'    )
```

```python
## Get the Column headers to prepare JSON
def get_column_names(schemas,ds_name,sorting_key='column_position'):
    column_Details=schemas[ds_name]
    columns=sorted(column_Details,key=lambda col:col[sorting_key])
    return [col['column_name'] for col in columns]
```

```python
#get the file names from file path
schemas=json.load(open(
    'F:/GitHub/PythonDataEngineering/PythonSQL/03_Postgre_Python/data-master/retail_db/schemas.json'
))
for file in src_file_names:
    file_path_list=(re.split('[//\\\]',file)) # covert file path into list  
    ds_name=file_path_list[-2] # here we will get the ds name ex orders
    file_name=file_path_list[-1]
    ds_columns=get_column_names(schemas,ds_name) # get the list of ds
    
    #Reading a file as a csv    
    df=pd.read_csv(file,names=ds_columns)
    #print(f'Shape of file {ds_name}/{file_name} is - {df.shape}')

    #output jsonfie path 
    # Writing this files to output  JSON dir
    op_dir='F:/GitHub/PythonDataEngineering/PythonSQL/03_Postgre_Python/data-master/retail_db_json'

    json_file_path=f'{op_dir}/{ds_name}/{file_name}'

    os.makedirs(f'{op_dir}/{ds_name}',exist_ok=True)
    # Write files to op folder
    df.to_json(json_file_path,orient='records',lines=True)

    
```

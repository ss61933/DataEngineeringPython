```python
type(open(path).read()) 
## So we do not use read this way , we use pandas lib or spark in read world
## this will bring the data in a memory and now its available for processing
```

```python
orders_str=open(path).read()
```

```python
# create a list using this data
orders_record=orders_str.splitlines()
```

```python
len(orders_record)
```

```python
orders_record[:10]
```

```python
# filter the orders for complete status
list(filter((lambda orders:orders.endswith ('CLOSED')),orders_record))
```

```python
# Row level transformation-> filter the orders for complete status
orders_f=list(filter((lambda orders:orders.split(',')[3] in('CLOSED','COMPLETE')),orders_record))
len(orders_f)
```

```python
# Row level transformation - > get unqiue statuses - since sets are unqiue we will wrap the 
#the op from list funtion in set
orders_u=set(map((lambda orders:orders.split(',')[3]),orders_record))
orders_u
```

```python
# sort data based on id
sortred_o=sorted(orders_record,key= lambda orders_record:orders_record.split(',')[2])
sortred_o
```
